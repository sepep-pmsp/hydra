```python
from hydra.resources.censo import CensoResource
import requests
import zipfile
import io
import pandas as pd
import csv
import os
```

```python
#Utilizando interface para baixar o resource
```

```python
class NewCensoResource(CensoResource):
    URL:str = 'https://ftp.ibge.gov.br/Censos/Censo_Demografico_2022/Agregados_por_Setores_Censitarios_preliminares/agregados_por_setores_csv/BR/Agregados_preliminares_por_setores_censitarios_BR.zip'
```

```python
class OldCensoResource(CensoResource):
    URL:str = 'https://ftp.ibge.gov.br/Censos/Censo_Demografico_2010/Resultados_do_Universo/Agregados_por_Setores_Censitarios/SP_Capital_20231030.zip'
```

```python
censo_new = NewCensoResource()
```

```python
censo_old = OldCensoResource()
```

```python
#Func para abrir como df o censo/ver
```

```python
def zip_to_dataframe(censo):
    censo_zip = censo.download_zipfile()
    zip_file = zipfile.ZipFile(io.BytesIO(censo_zip))
    csv_filename = zip_file.namelist()[0]
    
    if os.path.exists(csv_filename):
        print(f"O arquivo '{csv_filename}' já existe no diretório.")
        return
    
    # Extrair o arquivo CSV para o diretório atual
    zip_file.extract(csv_filename)
    
    # Ler o arquivo CSV como DataFrame do Pandas
    df = pd.read_csv(csv_filename, sep=";", quoting=csv.QUOTE_ALL)
    
    return df
```

```python
df_old = zip_to_dataframe(censo_old)
```

```python
df_new = zip_to_dataframe(censo_new)
```
